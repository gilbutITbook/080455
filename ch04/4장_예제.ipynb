#벡터 데이터베이스 크로마(Chroma) 데이터베이스 사용해보기

1. 설치

```python
pip install chromadb
```

2. 크로마 클라이언트 생성

```python
import chromadb
chroma_client = chromadb.Client() #크로마 데이터베이스 클라이언트 생성
```

3. 컬렉션 만들기

```python
collection = chroma_client.create_collection(name="my_collection") #컬렉션 생성
```

4. 컬렉션에 텍스트 document 추가

```python
collection.add(
    documents=[	#document에 저장할 데이터 입력
        "This is a document about pineapple",
        "This is a document about oranges",
        "Hello",
        "My name is chroma"
    ],
    ids=["id1", "id2", "id3", "id4"] #데이터의 index 입력
)
```

5. 데이터 조회

```python
results = collection.query(
    query_texts=["This is a query document about hawaii"], # 조회할 쿼리 입력
    n_results=2 # 조회될 결과 개수 입력
)
print(results)
```

6. 다른 질문으로 데이터 조회

```python
results = collection.query(
    query_texts=["What is your name?"], # 조회할 쿼리 입력
    n_results=1 # 조회될 결과 개수 입력
)
print(results)
```

#4_2_3 기본 예제 실습

1. 패키지 설치

```python
%pip install openai
```

2. 권한 인증 (개인 API key 입력)


```python
from openai import OpenAI

client = OpenAI(
    api_key="sk-" # 생성된 개인 API Key 입력
)
```

3. OpenAI API 호출

```python
def create_chat_completion(system_input, user_input, model="gpt-4o-mini", temperature=1.15, max_tokens=300):
    try:
        # 메시지 목록을 자동으로 생성합니다.
        messages = [
            {"role": "system", "content": system_input},
            {"role": "user", "content": user_input}
        ]

        # 모델 선택 및 요청
        response = client.chat.completions.create(
            model=model,
            messages=messages,
            temperature=temperature,  # 결과물의 randomness 조절, 범위: 0~2, 일정한 답변: 0.3, 창의적 결과 : 0.7
            # 항상 동일한 결과를 보장하지 않습니다.
            max_tokens=max_tokens  # 최대 토큰 수를 지정
        )
        # 생성된 응답을 반환합니다.
        return response
    except Exception as e:
        return f"Error: {str(e)}"

# 메시지 목록 예시
system_input = "너는 AI 분야의 최고전문가야."
user_input = "LLM으로 무엇을 할 수 있는지 설명해줘."

# API 호출 및 결과 출력
responses = create_chat_completion(system_input, user_input)
print(responses.choices[0].message.content)
```

#4_2_4 기본 예제 실습

1. 패키지 설치

```python
import google.generativeai as genai
```

2. 권한 인증 (개인 API key 입력)


```python
genai.configure(api_key="…")  # 생성한 API Key 입력
```

3. OpenAI API 호출

```python
def create_chat_completion(system_input, user_input, model="gemini-2.0-flash", temperature=1.0, max_tokens=300):
    try:
        # 제미나이에서는 system 프롬프트를 모델에게 직접 입력하는 방식으로 처리
        full_prompt = f"{system_input}\n\n사용자: {user_input}"

        # 모델 선택 및 요청
        response = genai.GenerativeModel(model).generate_content(
            full_prompt,
            generation_config={
                "temperature": temperature,
                "max_output_tokens": max_tokens
            }
        )

        # 생성된 응답을 반환합니다.
        return response.text
    except Exception as e:
        return f"Error: {str(e)}"

# 메시지 목록 예시
system_input = "너는 AI 분야의 최고 전문가야."
user_input = "LLM으로 무엇을 할 수 있는지 설명해줘."

# API 호출 및 출력
response = create_chat_completion(system_input, user_input)
print(response)
```
