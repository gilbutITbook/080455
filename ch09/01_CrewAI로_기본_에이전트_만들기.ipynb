### Agent ì •ì˜

```python
import os

from google.colab import userdata
os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')
```

```python
from crewai import Agent, Task, Crew, LLM

#ëª©ì°¨ ì„¤ì • ì—ì´ì „íŠ¸
outline_generator = Agent(
    role='Outline Generator',
    goal='ì£¼ì œì— ëŒ€í•œ ëª©ì°¨ë¥¼ ì‘ì„±í•´ ì¤˜.',
    llm = LLM(model = "openai/gpt-4o-mini", max_tokens = 1000),
    backstory='ë„ˆëŠ” ì •ë³´ë¥¼ ì²´ê³„ì ìœ¼ë¡œ ì •ë¦¬í•˜ê³  ë‹¤ì–‘í•œ ì£¼ì œì— ëŒ€í•œ í¬ê´„ì ì¸ ëª©ì°¨ë¥¼ ì‘ì„±í•˜ëŠ” ì „ë¬¸ê°€ì•¼.'
)
#ë³¸ë¬¸ ì‘ì„± ì—ì´ì „íŠ¸
writer = Agent(
    role='Writer',
    goal='ì£¼ì œì— ëŒ€í•´ ì—°êµ¬í•´ì„œ í¥ë¯¸ë¡œìš´ ê¸€ì„ ì¨ ì¤˜.',
    llm = LLM(model = "openai/gpt-4o-mini", max_tokens = 3000),
    backstory='ë„ˆëŠ” ë³µì¡í•œ ì •ë³´ë¥¼ ì½ê¸° ì‰¬ìš´ ê¸€ë¡œ ë°”ê¿€ ìˆ˜ ìˆëŠ” ëŠ¥ìˆ™í•œ ì‘ê°€ì•¼.'
)

# agent : 2ê°œ (ëª©ì°¨ ì „ìš© ì—ì´ì „íŠ¸, ë³¸ë¬¸ ì‘ì„± ì „ìš© ì—ì¸ì „íŠ¸)
```

### Task ì •ì˜

```python
# Task ì •ì˜
# Task í•„ìˆ˜ìš”ì†Œ (description, agent, expected_output)
outline_task = Task(
    description='ì¸ê°„ê³¼ AIì˜ ê³µì¡´ì— ëŒ€í•œ ê¸€ì˜ ëª©ì°¨ë¥¼ ì‘ì„±í•´ ì¤˜.',
    agent=outline_generator,
    expected_output="ì¸ê°„ê³¼ AIì˜ ê³µì¡´ì„ í¬ê´„ì ìœ¼ë¡œ ë‹¤ë£¨ëŠ” ëª©ì°¨"
)

writing_task = Task(
    description='ì—°êµ¬ ê²°ê³¼ì— ëŒ€í•œ ê¸°ì‚¬ë¥¼ ì‘ì„±í•´ ì¤˜.',
    agent=writer,
    expected_output='ì¸ê°„ê³¼ AIì˜ ê³µì¡´ì„ ë‹¤ë£¬ í¥ë¯¸ë¡œìš´ ê¸°ì‚¬'
)
```

### Crew ì •ì˜

```python
# Crew ì •ì˜
from crewai.process import Process

ai_impact_crew = Crew(
    agents=[outline_generator, writer],
    tasks=[outline_task, writing_task],
    verbose=True,
    Process=Process.sequential # ìˆœì°¨ì  ì‘ì—…ìˆ˜í–‰
)
```

```python
# Crew ì‹¤í–‰
result = ai_impact_crew.kickoff()
```
