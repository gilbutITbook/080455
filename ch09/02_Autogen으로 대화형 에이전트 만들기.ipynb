```python
import os
import warnings
warnings.filterwarnings('ignore')

from google.colab import userdata
os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')
```

```python
# OpenAI Autogen 패키지에서 ConversableAgent(대화형 에이전트) 클래스를 가져옵니다.
from autogen import ConversableAgent

# --------------------------------------------------------
# 1. Junior_Coder_Agent 생성 (초급 개발자)
# --------------------------------------------------------
coder_agent = ConversableAgent(
    name="Junior_Coder_Agent",  # 에이전트의 이름을 설정
    system_message="""너는 1년차 파이썬 프로그램머야. 모르는 것은 Senior Coder Agent에게 물어봐""",

    llm_config={
        "config_list": [
            {
                "model": "gpt-4o-mini",
                "api_key": os.environ["OPENAI_API_KEY"],  # API 키를 사용하여 OpenAI와 연결
            }
        ]
    }
)

# --------------------------------------------------------
# 2. Senior_Coder_Agent 생성 (고급 개발자)
# --------------------------------------------------------
manager_agent = ConversableAgent(
    name="Senior_Coder_Agent",  # 에이전트의 이름을 설정
    system_message="""너는 20년차 파이썬 프로그래머야.
질문이 주어지면, 해당 질문에 전문성을 가지고 대답해줘.
만약 코드가 주어졌다면 주어진 코드를 검토하고 효율성을 높일 방안을 제시해줘""",

    llm_config={
        "config_list": [
            {
                "model": "gpt-4o",
                "api_key": os.environ["OPENAI_API_KEY"],  # API 키 사용
            }
        ]
    }
)

# --------------------------------------------------------
# 3. Junior_Coder_Agent가 Senior_Coder_Agent에게 질문하기
# --------------------------------------------------------
chat_result = coder_agent.initiate_chat(
    manager_agent,  # 대화를 할 상대 (Senior_Coder_Agent)
    message="퀵 정렬 코드에 대해서 설명해주세요.",
    # Junior_Coder_Agent가 Senior_Coder_Agent에게 "퀵 정렬 코드에 대해서 설명해주세요"라는 질문을 던짐

    summary_method="reflection_with_llm",
    # 대화 요약 방법을 지정 (AI가 요약을 도와줌)

    max_turns=2,
    # 최대 2번까지 대화 가능 (질문-답변 2회)
)
```

```python
print(chat_result.summary)
```
